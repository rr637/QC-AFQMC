```python
import numpy as np
import os
```

```python
VQE_DIR = "VQE_Trials"


for ent in sorted(os.scandir(VQE_DIR), key=lambda e: e.name):
    if not (ent.is_file() and ent.name.lower().endswith(".npz")):
        continue
    print("\n=== File:", ent.name, "===")

    try:
        with np.load(ent.path, allow_pickle=True) as vqe_run_data:
            print("Final Energy:", vqe_run_data["energy_list"][-1])
            print("num_dets: ", len(vqe_run_data["coeffs"]))
            print("number of parameters:", vqe_run_data["num_params"])
            print("runtimes:",   vqe_run_data["runtimes"])

    except Exception as e:
        print("Error loading", ent.name, ":", e)
```

```python
AFQMC_DIR = f"VQE_Trials/AFQMC_H4_1.0_chain_beta2.0"


for ent in sorted(os.scandir(AFQMC_DIR), key=lambda e: e.name):
    if not (ent.is_file() and ent.name.lower().endswith(".npz")):
        continue
    print(ent.name)
    try:
        with np.load(ent.path, allow_pickle=True) as afqmc_run_data:
            
            print("Trial_energy: ", afqmc_run_data["trial_energy"])
            print("Number of Dets: ", afqmc_run_data["num_dets"])
            print("mean_ac: ", afqmc_run_data['mean_ac'])
            print("err_ac (Ha): ", afqmc_run_data['err_ac'])

    except Exception as e:
        print(f"Skipping {ent.name}: failed to load ({e})")
```
